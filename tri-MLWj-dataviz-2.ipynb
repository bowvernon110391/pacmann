# Soal 1A (20 poin)

- Buatlah visualisasi histogram dan analisis untuk variabel numerikal:
    * `SalePrice`
    * `GrLivArea`
    * `GarageArea`

- Terapkan pemrosesan data yang sesuai untuk masing-masing variabel.
- Pada tiap histogram mohon tampilkan pula informasi central tendecy dari tiap variable (mean/median/mode)
- Tampilkan histogram untuk ketiga variabel di atas dengan format subplot seperti berikut:

```python
df_train['GrLivArea'].mode()
```

```python
clean_sale_price=df_train.where(df_train['SalePrice'] < 450000)['SalePrice']
```

```python
from sklearn.cluster import KMeans

clustering = KMeans(n_clusters=2, n_init=2)
gr_liv_area = df_train['GrLivArea']
x = np.array(gr_liv_area)
clustering.fit(x.reshape(-1,1))

label = pd.Series(clustering.labels_)
data = pd.Series(x)
data = pd.concat([data, label], axis=1)
data.columns = ['GrLivArea', 'label']
data
```

```python
garage_area = df_train['GarageArea'].copy()
garage_area[garage_area < 125] = garage_area.median()
```

```python
fig, axs = plt.subplots(2, 3, figsize=(18, 10))

axs[0][1].set_title('Histogram For Non-Transformed Data')
axs[1][1].set_title('Histogram For Transformed Data')

cols = ['SalePrice', 'GrLivArea', 'GarageArea']

# cleansed


# plot non transformed
for (i,c) in enumerate(cols):
    sns.distplot(df_train[c], 20, ax=axs[0][i])
    mean = df_train[c].mean()
    median = df_train[c].median()
    mode = df_train[c].mode()[0]

    axs[0][i].axvline(mean, label='mean', color='r', linestyle='dashed')
    axs[0][i].axvline(median, label='median', color='g', linestyle='dashed')
    axs[0][i].axvline(mode, label='mode', color='b', linestyle='dashed')

    axs[0][i].legend(loc='upper right')

# plot transformed SalePrice
sns.distplot(clean_sale_price, 20, ax=axs[1][0])
axs[1][0].axvline(clean_sale_price.mean(), label='mean', color='r', linestyle='dashed')
axs[1][0].axvline(clean_sale_price.median(), label='median', color='g', linestyle='dashed')
axs[1][0].axvline(clean_sale_price.mode()[0], label='mode', color='b', linestyle='dashed')
axs[1][0].legend(loc='upper right')

# transform GrLivArea
c0 = data[data['label'] == 0]['GrLivArea']
c1 = data[data['label'] == 1]['GrLivArea']

sns.distplot(c0, 20, ax=axs[1][1])
sns.distplot(c1, 20, ax=axs[1][1])
axs[1][1].axvline(c0.mean(), color='r', label='mean_c0', linestyle='dashed')
axs[1][1].axvline(c0.median(), color='g', label='median_c0', linestyle='dashed')
axs[1][1].axvline(c0.mode()[0], color='b', label='mode_c0', linestyle='dashed')
axs[1][1].axvline(c1.mean(), color='r', label='mean_c1', linestyle='dashed')
axs[1][1].axvline(c1.median(), color='g', label='median_c1', linestyle='dashed')
axs[1][1].axvline(c1.mode()[0], color='b', label='mode_c1', linestyle='dashed')
axs[1][1].set_xlim([0, 3500])
axs[1][1].legend(loc='upper right')

#transform garage_area
sns.distplot(garage_area, 20, ax=axs[1][2])
axs[1][2].axvline(garage_area.mean(), label='mean', color='r', linestyle='dashed')
axs[1][2].axvline(garage_area.median(), label='mean', color='g', linestyle='dashed')
axs[1][2].axvline(garage_area.mode()[0], label='mode', color='b', linestyle='dashed')
axs[1][2].set_xlim([0, 1200])
axs[1][2].legend(loc='upper right')

plt.show()
```
